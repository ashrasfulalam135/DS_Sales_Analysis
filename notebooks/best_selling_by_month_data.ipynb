import necessary libraries

```python
import pandas as pd
```

Read data

```python
all_data = pd.read_csv('../data/procesed/all_months_processed_data.csv') 
```

Best Selling Month data

```python
all_data = all_data.groupby('Month').sum().reset_index()
all_data = all_data[['Month','Sales']]
all_data.to_csv("../data/final/sales_by_month.csv", index=False)
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,all_data['Sales'])
plt.xticks(months)
plt.xlabel('Month')
plt.ylabel('Sales in USD')
plt.savefig('../data/final/best_selling_month.png')
plt.show()
```
