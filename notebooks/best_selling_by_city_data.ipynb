import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Read data

```python
all_data = pd.read_csv('../data/procesed/all_months_processed_data.csv')
```

Best Selling Month data

```python
all_data = all_data.groupby('City').sum().reset_index()
all_data = all_data[['City','Sales']]
all_data.to_csv("../data/final/sales_by_city.csv", index=False)
```

Show data by bar chart

```python
plt.bar(all_data['City'],all_data['Sales'])
plt.xticks(all_data['City'], rotation='vertical', size=8)
plt.xlabel('Cities')
plt.ylabel('Sales in USD')
plt.savefig('../data/final/best_selling_city.png')
plt.show()
```
